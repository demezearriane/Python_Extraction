### Fonction qui recupere les liens des hotels

```python
def recup_lien():
    urlAccueil="https://www.booking.com/reviews/cm/city/yaounde.fr.html"
    response = get(urlAccueil)
    source = None # Le code source de la page
    
    if response.status_code == 200 :
        # Si la requete s'est bien passee
        source = response.text

    if source :
        # Si le code source existe
        selector = Selector(text=source)
        liens=selector.xpath('//ul/li/div/div/ul/li/a/@href').extract()
    
    for i in range(len(liens)):
        ListeLien.append("https://www.booking.com/"+liens[i])
    
    return(ListeLien)
```

```python
def Extract_bookingHotels(l):
  from requests import get
  from scrapy import Selector


  for lien in range(len(l)):
    url=l[lien]
    response = get(url) #Téléchargement
    source = None # Le code source de la page 
    if response.status_code == 200 :
      # Si la requete s'est bien passee
      source = response.text

    if source :
      # Si le code source existe
      selector = Selector(text=source)
      titles=selector.xpath('//ul/li/div/div/div//text()').getall()

    #for i in range(len(titles)):
     # print(titles[i])

    #Stockage des dans le fichier 'dataHotels.csv'
    return (titles)
```

### Fonction qui stocke les donnees dans un fichier et supprime les lignes vides

```python
def Stock_dataHotels(titles):
    import fileinput
    import sys
    
    with open('/home/arriane/Documents/ESSFAR/PYTHON 2/Projet Python2/dataHotels.csv', mode='w',encoding="utf-8") as fich:
        for i in range(len(titles)):
            fich.write("{}\n".format(titles[i]))
    
    #Supression des lignes vides
    for line_number, line in enumerate(fileinput.input('dataHotels.csv', inplace=1)):
        if len(line) > 1:
            sys.stdout.write(line)
```
